This notebook performs basic sales analysis and visualization.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('Sample-Superstore.csv')
df.head()
```

```
df['Revenue'] = df['Sales']
df.head()
```

```
total_revenue = df['Revenue'].sum()
total_revenue
```

```
top_products = df.groupby('Product Name')['Revenue'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='bar', title='Top 5 Products')
plt.show()
```

```
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.resample('M', on='Order Date')['Revenue'].sum()
monthly_sales.plot(title='Monthly Sales Trend')
plt.show()
```

## Business Insights
- Technology products generate higher revenue
- Sales increase in mid and end of the year
- West and East regions perform well
