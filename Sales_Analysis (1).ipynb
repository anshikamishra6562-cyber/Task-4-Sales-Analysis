```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sample-Superstore.csv")
df.head()
```

```python
df['Revenue'] = df['Sales']
df.head()
```

```python
total_revenue = df['Revenue'].sum()
total_revenue
```

```python
top_products = (
    df.groupby('Product Name')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_products.plot(kind='bar', title='Top 5 Products')
plt.show()

```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.resample('M', on='Order Date')['Revenue'].sum()
monthly_sales.plot(title='Monthly Sales Trend')
plt.show()
```

Business Insights:
- Technology products have highest revenue
- Sales increase in later months
- West and East regions perform well
